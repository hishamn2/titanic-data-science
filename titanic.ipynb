```python
df["Age"]=df["Age"].fillna(df["Age"].median)
df["Fare"]=df["Fare"].fillna(df["Fare"].mean)
df=df.drop(columns=["Cabin"])
print(df.isnull().sum())
```

```python
# Quick stats
print(df.describe())

# Look at categorical counts
print(df['Sex'].value_counts())
print(df['Embarked'].value_counts())
print(df['Pclass'].value_counts())
```

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')
print(df.isnull().sum())
```

```python
df = pd.read_csv("tested.csv")  # reload clean
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
# Histogram of Age
# Simple histograms
df['Age'].hist(bins=20, grid=True, figsize=(10,5),density=True)
plt.show()
df['Fare'].hist(bins=20, grid=True, figsize=(10,5))
plt.show()
```

```python
plt.scatter(df["Fare"],df["Age"],c="black",s=20,alpha=0.3)
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("fare vs age scatter plot")
plt.show()
```
